```python
import pandas as pd

# CSV 파일 읽기
test_csv = pd.read_csv("/home/aicontest/construct/data/test.csv")
answer_csv = pd.read_csv("/home/aicontest/construct/experiment/result/baseline_submission_RAG.csv")

# test_csv의 각 행을 순회합니다.
for idx, row in test_csv.iterrows():
    # '사고원인'과 'ID' 값을 가져옵니다.
    question = row['사고원인']
    id_val = row['ID']

    # answer_csv에서 ID가 일치하는 행을 찾습니다.
    answer_row = answer_csv[answer_csv['ID'] == id_val]

    if not answer_row.empty:
        # 예를 들어 'answer'라는 컬럼에 답변이 있다고 가정합니다.
        answer = answer_row.iloc[0]['재발방지대책 및 향후조치계획']
    else:
        answer = "해당 ID에 해당하는 답변을 찾을 수 없습니다."

    print(f"question : {question} \n answer : {answer}")
```
